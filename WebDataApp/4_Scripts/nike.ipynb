```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
navegador = webdriver.Chrome()
```

```python
lista_produtos = []
for j in range(0,50):
    navegador.get(f'https://www.nike.com.br/nav/genero/masculino/idade/adulto/tipodeproduto/roupas?page={j}')    
    for i in range(0,31):
        try:
            print (navegador.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/div[1]/div[3]/div[2]/div[2]/div[1]/div[{i}]/div/a/div[2]').text)
            avalia = navegador.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/div[1]/div[3]/div[2]/div[2]/div[1]/div[{i}]/div/a/div[2]').text
            lista_produtos(avalia)
        except:
            pass
        try:
            print (navegador.find_element(By.XPATH,f'/html/body/div[1]/div/div/div/div[1]/div[3]/div[2]/div[2]/div[1]/div[{i}]/a/div[2]').text)
            avalia2 = navegador.find_element(By.XPATH,f'/html/body/div[1]/div/div/div/div[1]/div[3]/div[2]/div[2]/div[1]/div[{i}]/a/div[2]').text
            lista_produtos.append(avalia2)
        except:
            pass
```

```python
lista_produtos2 = [tuple(x.split('\n')) for x in lista_produtos]
```

```python
dfnike = pd.DataFrame(lista_produtos2, columns=["Nome", "Modelo","Valor c desc", "Valor s desc", "Porcentagem desc", "Avaliacao"," "])
```

```python
dfnike.to_csv('../0_Bases_originais/nikebruto.csv', sep= ';', index= False)
```
